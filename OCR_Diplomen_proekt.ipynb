Изчерване на кутии при срещнат текст

```python
def draw_boxes(image, bounds, color='red', width=2):
    draw = ImageDraw.Draw(image)
    #font_type = ImageFont.truetype('LiberationSerif-Regular.ttf', 20)

    for bound in bounds:
        p0, p1, p2, p3 = bound[0]
        #print(bound[0])
        draw.line([*p0, *p1, *p2, *p3, *p0], fill=color, width=width)
        #draw.text(xy=(*p0, *p1), text= bound[1], fill=(255,69,0), font=font_type)    

    return image
```
